```python
# Create lsoa dataframe
lsoa_df = pd.read_csv('lsoa_bike.csv', index_col = 0)
lsoa_df.dropna(inplace = True) 
lsoa_df.head()
```

```python
# Store latitude and longitude in locations
locs6 = lsoa_df[["LAT_y", "LONG__y"]]

# Fill NaN values and convert to float
incedents6 = lsoa_df["0"].astype(float)
```

```python
# Convert IMD Score to a list

IMD = lsoa_df["LSOA IMD Score_y"].tolist()
# Create a map using coordinates to set markers
marker_locations = lsoa_df[["LAT_y", "LONG__y"]]
```

```python
# Customize the size of the figure
figure_layout = {
    'width': '800px',
    'height': '600px',
    'padding': '1px',
    'margin': '0 auto 0 auto'
}
fig6 = gmaps.figure(layout=figure_layout)

markers = gmaps.symbol_layer(
    marker_locations, fill_color='rgba(0, 0, 255, 0.1)',
    stroke_color='rgba(0, 0, 255, 0.1)', scale=2,
    info_box_content=[f"IMD Score: {score}" for score in IMD]
)

fig6.add_layer(markers)

# Create heat layer
heat_layer = gmaps.heatmap_layer(locs6, weights=incedents6, 
                                 dissipating=False, max_intensity=10,
                                 point_radius=.01)


# Add layer
fig6.add_layer(heat_layer)

# Display figure
fig6
```
