```python
import gmaps
import pandas as pd

# Google developer API key
from config import gkey

# Configure gmaps
gmaps.configure(api_key=gkey)
```

```python
# Create crime dataframe
crime_df = pd.read_csv('clean_2019_1.csv')
crime_df.dropna(inplace = True) 
crime_df.head()
```

```python
new_index_df = crime_df.set_index("Crime type")
new_index_df.head()
```

```python
bike_locs_df = new_index_df.loc[["Bicycle theft"], ["Latitude",'Longitude']]
bike_locs_df.head()
```

```python
# Customize the size of the figure
figure_layout = {
    'width': '800px',
    'height': '600px',
    'padding': '1px',
    'margin': '0 auto 0 auto'
}
fig = gmaps.figure(layout=figure_layout)
```

```python
# Store latitude and longitude in locations
locations = bike_locs_df[["Latitude", "Longitude"]]
```

```python
# Add coordinates as a layer
markers = gmaps.marker_layer(locations)


# Add the layer to the map
fig.add_layer(markers)

# Print map
fig
```
