```python
import gmaps
import pandas as pd

# Google developer API key
from config import gkey

# Configure gmaps
gmaps.configure(api_key=gkey)
```

```python
# Create crime dataframe
crime_df = pd.read_csv('clean_2019_1.csv')
crime_df.dropna(inplace = True) 
crime_df.head()
```

```python
new_index_df = crime_df.set_index("Crime type")
new_index_df.head()
```

```python
bike_locs_df = new_index_df.loc[["Bicycle theft"], ["Latitude",'Longitude']]
bike_locs_df.head()
```

```python
# Customize the size of the figure
figure_layout = {
    'width': '800px',
    'height': '600px',
    'padding': '1px',
    'margin': '0 auto 0 auto'
}
fig = gmaps.figure(layout=figure_layout)
```

```python
# Store latitude and longitude in locations
locations = bike_locs_df[["Latitude", "Longitude"]]
```

```python
# Add coordinates as a layer
markers = gmaps.marker_layer(locations)


# Add the layer to the map
fig.add_layer(markers)

# Print map
fig
```

```python
# Create LSOA co-ordinates dataframe
LSOA_df = pd.read_csv('Lower_Layer_Super_Output_Areas__December_2011__Boundaries_Full_Clipped__BFC__EW_V3.csv')
LSOA_df.dropna(inplace = True) 
LSOA_df.head()
```

```python
LSOA_df1 = LSOA_df.rename(columns={"LSOA11CD":"LSOA code"})
LSOA_df1
```

```python
combinedx = pd.merge(crime_df, LSOA_df1, on = "LSOA code")
combinedx.head()
```

```python
combinedx.columns
```

```python
West_midlands_dfx = combinedx.reindex(['Crime ID', 'Crime type', 'Last outcome category', 'Month', 'Longitude',
       'Latitude', 'Location', 'LSOA code', 'LSOA name', 'LONG_', 'LAT', 'LSOA IMD Score',
       'LSOA IMD Rank', 'LSOA Income Rank', 'District code (2019)',
       'District name (2019)', 'Description', 'Population',
       'Index of Multiple Deprivation (IMD)',
       'Income deprivation- Average score ',
       'Income deprivation - Rank of average score ',
       'Income - Rank of proportion of LSOAs in most deprived 10% nationally '
       ], axis=1)
```
