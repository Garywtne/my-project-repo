```python
# Create bike theft dataframe
bike_theft_df = pd.read_csv('bike_theft.csv', index_col = 0)
bike_theft_df.dropna(inplace = True) 
bike_theft_df.head()
```

```python
# Store latitude and longitude in locations
locations = bike_theft_df[["LAT", "LONG_"]]

# Fill NaN values and convert to float
thefts = bike_theft_df["Bicycle theft"].astype(float)
```

```python
# Customize the size of the figure
figure_layout = {
    'width': '800px',
    'height': '600px',
    'padding': '1px',
    'margin': '0 auto 0 auto'
}
fig1 = gmaps.figure(layout=figure_layout)



# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=thefts, 
                                 dissipating=False, max_intensity=20,
                                 point_radius=.01)


# Add layer
fig1.add_layer(heat_layer)

# Display figure
fig1
```
